## Get Historical Data

```python
import requests
i = 0
time = 1535731200
LA = []
while i < 30:    
    time = 1535731200 + i * 60 * 60 * 24 
    url = 'https://api.darksky.net/forecast/ae46a27eae55d5947baedcc7af1e12af/34.052235,-118.243683,' + str(time)
    #print(url)
    r = requests.get(url)
    high = (r.json().get('daily').get('data'))[0].get('apparentTemperatureHigh')
    low = (r.json().get('daily').get('data'))[0].get('apparentTemperatureLow')
    mean_tep = round((high + low) / 2,2)
    max_tep = (r.json().get('daily').get('data'))[0].get('apparentTemperatureMax')
    min_tep = (r.json().get('daily').get('data'))[0].get('apparentTemperatureMin')
    LA.append({'city:':'Los Angeles','temperature':mean_tep,'maximum temperature':max_tep,'minimum temperature':min_tep})
    i += 1
print(LA)
```

```python
import requests
from bs4 import BeautifulSoup
import time

url = 'http://www.showmeboone.com/sheriff/JailResidents/JailResidents.asp'
response = requests.get(url)
html = response.content.decode('ascii')

soup = BeautifulSoup(html, "lxml")
#print(soup)
#print(soup.prettify())

start = time.time()

last_name_td_list = soup.find_all('td', attrs = {"data-th": "Last Name"})
first_name_td_list = soup.find_all('td', attrs = {"data-th": "First Name"})
age_td_list = soup.find_all('td', attrs = {"data-th": "Age"})

i = 0
lastname = []
firstname = []
age = []

while i < len(last_name_td_list):
    lastname.append(last_name_td_list[i].string)
    firstname.append(first_name_td_list[i].string)
    age.append(age_td_list[i].string)
    i += 1

urln = []
personal = soup.find_all ('a' , attrs = {'class' : "_lookup btn btn-primary" })

for j in personal:
     id1 = j.get('href')
     urln.append('https://report.boonecountymo.org/mrcjava/servlet/'+id1)
#print(urln)

i = 0
#responsen = []
#detailn = []
#soupn = []
case = []
bail = []

while i < len(urln):
    responsen = requests.get(urln[i])
    detailn = responsen.content.decode('ascii')
    soupn = BeautifulSoup(detailn,"lxml")
    case_td_list = soupn.find_all('td', attrs = {"data-th": "CASE #"})
    case.append(len(case_td_list))
    bailn = 0
    for x in soupn.find_all ('td' , attrs = {'data-th' : 'BAIL'}):
        string = x.get_text().strip('\n').replace(',', '').strip()
        if string == '':
            string = 0
        #print(string)
        else:
            string = float(string)
        #print(string)
        bailn = bailn + string
    bail.append(bailn)
    i += 1

end = time.time()
time1 = end - start

print('Last Name','First Name','Age','Total Cases','Total Bail')
i = 0
while i < len(last_name_td_list):
    print(lastname[i],firstname[i],age[i],case[i],bail[i])
    i += 1

print(time1)
```

## Threading

```python
import threading
import requests
from bs4 import BeautifulSoup
import time

start = time.time()

url = 'http://www.showmeboone.com/sheriff/JailResidents/JailResidents.asp'
response = requests.get(url)
html = response.content.decode('ascii')
soup = BeautifulSoup(html, "lxml")

def mainpage():
    last_name_td_list = soup.find_all('td', attrs = {"data-th": "Last Name"})
    first_name_td_list = soup.find_all('td', attrs = {"data-th": "First Name"})
    age_td_list = soup.find_all('td', attrs = {"data-th": "Age"})

    i = 0
    lastname = []
    firstname = []
    age = []

    while i < len(last_name_td_list):
        lastname.append(last_name_td_list[i].string)
        firstname.append(first_name_td_list[i].string)
        age.append(age_td_list[i].string)
        i += 1
    #print(lastname,firstname,age)

t1 = Thread(target = mainpage,args = ())
t1.start()

urln = []
personal = soup.find_all ('a' , attrs = {'class' : "_lookup btn btn-primary" })

for j in personal:
    id1 = j.get('href')
    urln.append('https://report.boonecountymo.org/mrcjava/servlet/'+id1)

tLock = threading.Lock() 
case = [] 
bail = [] 

def personalpage(urlnn):
    #url1 = urln[i]
    responsen = requests.get(urlnn)
    detailn = responsen.content.decode('ascii')
    soupn = BeautifulSoup(detailn,"lxml")
    case_td_list = soupn.find_all('td', attrs = {"data-th": "CASE #"})
    casen = len(case_td_list)
    bailn = 0
    for x in soupn.find_all ('td' , attrs = {'data-th' : 'BAIL'}):
        string = x.get_text().strip('\n').replace(',', '').strip()
        if string == '':
            string = 0
        #print(string)
        else:
            string = float(string)
        #print(string)
        bailn = bailn + string
    #print(case)
    #print(bail)
    tLock.acquire()
    case.append(casen)
    #print(case)
    #bail[i] = bailn
    bail.append(bailn)
    #print(bail)
    tLock.release()

i = 0
while i < len(urln):
    #urlnn = urln[i]
    t2 = Thread(target = personalpage,args = (urln[i],))
    t2.start()
    t2.join()
    i += 1

end = time.time()
time2 = end - start

print('Last Name','First Name','Age','Total Cases','Total Bail')

i = 0
while i < len(last_name_td_list):
    print(lastname[i],firstname[i],age[i],case[i],bail[i])
    i += 1

print(time2)
```

## Plot

```python
import matplotlib.pyplot as plt
i = 0
while i < len(urln):
    responsen = requests.get(urln[i])
    detailn = responsen.content.decode('ascii')
    soupn = BeautifulSoup(detailn,"lxml")
    charge_td_list = soupn.find_all('td', attrs = {"data-th": "CHARGE DESCRIPTION"}) 
    #print(charge_td_list)
    j = 0
    while j < len(charge_td_list):
        charge.append(charge_td_list[j].string)
        j += 1
    i += 1
#print(charge)

charge_counter = {}
i = 0
while i < len(charge):
    if charge[i] in charge_counter:
        charge_counter[charge[i]] += 1
    else:
        charge_counter[charge[i]] = 1
    i += 1
popular_charge = sorted(charge_counter, key = charge_counter.get, reverse = True)
 
top_10 = popular_charge[:10]

#top_10

p = 0
count = [0] * len(top_10)
while p < len(top_10):
    q = 0
    while q < len(charge):
        if top_10[p] == charge[q]:
            count[p] += 1
            q += 1
        else:
            q += 1
    count[p] = count[p] / len(charge)
    p += 1

labels = top_10
sizes = count

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True)
 
plt.axis('equal')
plt.show()
```
